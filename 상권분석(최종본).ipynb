# 이상치 확인 및 처리

```python
typeOfList = df_commercial['서비스_업종_코드_명'].unique()
for item in typeOfList :
  length = len(df_commercial.loc[df_commercial['서비스_업종_코드_명'] == item])
  print('Item Name : {}, Item Length : {}'.format(item, length))
```

```python
# 이상치 제거 - 학원, 고시원, 편의점, 문구점, 패스트푸드점 - 비지니스 모델에 해당하지 않는 서비스 업종 Row를 제거함
CommercialRemoveItems = ['학원', '고시원', '편의점', '문구점', '패스트푸드점', '슈퍼마켓', '전자상거래업', 
                         '스포츠 강습', '스포츠클럽', '일반의원', '치과의원', '한의원', '부동산중개업']

privateList = df_commercial.loc[(df_commercial['서비스_업종_코드_명'].str.contains('학원'))].index.tolist()

df_commercial = df_commercial.drop(privateList)

# 적합하지 않은 업종에 대한 Row를 제거함
for item in CommercialRemoveItems : 
  if item == '학원' : 
    idxList = df_commercial.loc[(df_commercial['서비스_업종_코드_명'].str.contains('학원'))].index.tolist()
  else :
    idxList = df_commercial.loc[df_commercial['서비스_업종_코드_명'] == item].index.tolist()
  df_commercial = df_commercial.drop(idxList)
  df_commercial = df_commercial.reset_index(drop=True)

print("len", len(df_commercial['서비스_업종_코드_명'].unique()))
df_commercial['서비스_업종_코드_명'].unique()
```

```python
# 20 ~ 40 사이의 매출건수의 합계의 평균을 구하는 로직
codeList = sorted(df_commercial['서비스_업종_코드'].unique())
avgDict = {}
print('코드명', df_commercial['서비스_업종_코드_명'].unique())
for code in codeList :
  is_code = df_commercial['서비스_업종_코드'] == code
  filter_df_commercial = df_commercial[is_code]
  newDict = {}
  newDict[code] = round((filter_df_commercial['연령대_20_매출_건수'].mean() + filter_df_commercial['연령대_30_매출_건수'].mean() + filter_df_commercial['연령대_40_매출_건수'].mean()) / 3, 2) 
  avgDict.update(newDict)

avgDict
```

```python
# 20, 30, 40 매출 건수를 더해 각 서비스 업종 코드별 평균치 보다 아래의 Row index를 구함
removeIdxList = []

for key in avgDict.keys() :
  df_service_code = df_commercial[df_commercial['서비스_업종_코드'] == key]
  sum = (df_service_code['연령대_20_매출_건수'] + df_service_code['연령대_30_매출_건수'] + df_service_code['연령대_40_매출_건수']) / 3
  sumList = sum.tolist()
  idxList = sum.index.tolist()
  for i in range(len(sumList)) :
    if sumList[i] < avgDict.get(key):
      removeIdxList.append(idxList[i])

len(removeIdxList)
```

```python
# 20, 30, 40 연령 매출 건수가 모두 0 인 Row의 index를 구함
rmZeroList = []
sum = df_commercial['연령대_20_매출_건수'] + df_commercial['연령대_30_매출_건수'] + df_commercial['연령대_40_매출_건수']
sumList = sum.tolist()
idxList = sum.index.tolist()
for idx in range(len(sumList)): 
  if sumList[idx] == 0 :
    rmZeroList.append(idxList[idx])

len(rmZeroList)
```

# 관계분석을 위한 시각화

```python
# 1) 20~40대 각각 업종별 매출금액 내림차순으로 상위 10개씩으로만 정렬 부탁드립니다. (총 그래프 3개)
# 1 ~ 3 그래프는 상위 10 4 - 6 그래프는 하위 10
# 단위는 1E6은 1 x 10을 나타냄. 1백만 11은 10조 9는 1천억
filterList = ['서비스_업종_코드_명']
ageList = ['연령대_20_매출_건수', '연령대_30_매출_건수', '연령대_40_매출_건수']

gb_df_commercial = df_commercial.groupby(df_commercial['서비스_업종_코드_명'])
gb_sum = gb_df_commercial.sum().loc[:, ageList]
gb_sum.insert(0, '서비스_업종_코드_명', list(gb_df_commercial.groups))

for age in ageList :
  df_sort = gb_sum.sort_values(age, ascending=[False])
  df_top = df_sort.head(10)
  plt.figure(figsize = (8,5))
  plt.xticks(rotation = - 45 )
  sns.barplot(x='서비스_업종_코드_명', y=age, data=df_top, color="b")

for age in ageList :
  df_sort = gb_sum.sort_values(age, ascending=[True])
  df_bottom = df_sort.head(10)
  plt.figure(figsize = (8,5))
  plt.xticks(rotation = - 45 )
  sns.barplot(x='서비스_업종_코드_명', y=age, data=df_bottom, color="b")
```

```python
# 2) 20~40대 각각 지역구별 매출 금액 내림차순으로 상위, 하위 10개구만 표시되게 정렬
# 1 ~ 3 그래프는 상위 10 4 - 6 그래프는 하위 10
# 단위는 1E6은 1 x 10을 나타냄. 1백만 11은 10조 9는 1천억
filterList = ['구명']
ageList = ['연령대_20_매출_건수', '연령대_30_매출_건수', '연령대_40_매출_건수']

gb_df_commercial = df_commercial.groupby(df_commercial['구명'])
gb_sum = gb_df_commercial.sum().loc[:, ageList]
gb_sum.insert(0, '구명', list(gb_df_commercial.groups))
  
for age in ageList :
  df_sort = gb_sum.sort_values(age, ascending=[False])
  df_top = df_sort.head(10)
  plt.figure(figsize = (8,5))
  plt.xticks(rotation = - 45 )
  sns.barplot(x='구명', y=age, data=df_top, color="b")

for age in ageList :
  df_sort = gb_sum.sort_values(age, ascending=[True])
  df_bottom = df_sort.head(10)
  plt.figure(figsize = (8,5))
  plt.xticks(rotation = - 45 )
  sns.barplot(x='구명', y=age, data=df_bottom, color="b")
```

```python
# 시간대 매출 건수로 비율을 계산하여 정수형으로 변환 후 넣어준다
df_nameList = df_clean_commercial.loc[:, ["구명", '서비스_업종_코드']]
df_ratio_sex = df_clean_commercial.iloc[:, 72:74].div(df_clean_commercial.iloc[:, 72:74].sum(axis=1), axis=0)
df_ratio_sale = df_clean_commercial.iloc[:, 66:72].div(df_clean_commercial.iloc[:, 66:72].sum(axis=1), axis=0)

df_ratio_sex['남성_매출_비율'] = round(df_ratio_sex['남성_매출_건수'] * 100)
df_ratio_sex['여성_매출_비율'] = round(df_ratio_sex['여성_매출_건수'] * 100)

df_ratio_sex_to_int = df_ratio_sex.astype(int, errors='ignore')
df_ratio_sex = df_ratio_sex_to_int.drop(['남성_매출_건수', '여성_매출_건수'], axis=1)

df_ratio_sale['시간대_00~06_매출_비율'] = round(df_ratio_sale['시간대_건수~06_매출_건수'] * 100)
df_ratio_sale['시간대_06~11_매출_비율'] = round(df_ratio_sale['시간대_건수~11_매출_건수'] * 100)
df_ratio_sale['시간대_11~14_매출_비율'] = round(df_ratio_sale['시간대_건수~14_매출_건수'] * 100)
df_ratio_sale['시간대_14~17_매출_비율'] = round(df_ratio_sale['시간대_건수~17_매출_건수'] * 100)
df_ratio_sale['시간대_17~21_매출_비율'] = round(df_ratio_sale['시간대_건수~21_매출_건수'] * 100)
df_ratio_sale['시간대_21~24_매출_비율'] = round(df_ratio_sale['시간대_건수~24_매출_건수'] * 100)
df_ratio_sale_to_int = df_ratio_sale.astype(int, errors='ignore')
df_ratio_sale = df_ratio_sale_to_int.drop(['시간대_건수~06_매출_건수',
       '시간대_건수~11_매출_건수', '시간대_건수~14_매출_건수', '시간대_건수~17_매출_건수',
       '시간대_건수~21_매출_건수', '시간대_건수~24_매출_건수'], axis=1)

df_clean_drop = pd.concat([df_nameList, df_ratio_sale, df_ratio_sex], axis=1)
df_clean_drop
```
